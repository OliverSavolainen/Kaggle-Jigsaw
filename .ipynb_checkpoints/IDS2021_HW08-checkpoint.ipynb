```python
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
train = pd.read_csv('traindata.csv') 
test = pd.read_csv('testdata.csv') 
train.loan = train.loan.map(dict(yes=1, no=0)) 
train.default = train.default.map(dict(yes=1, no=0)) 
train.housing = train.housing.map(dict(yes=1, no=0)) 
co=train.select_dtypes(include='object').columns 
train_dum = pd.get_dummies(train,columns=co)
X = train_dum.drop(columns=['y']) 
y = train_dum['y'] 
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=45)
test.loan = test.loan.map(dict(yes=1, no=0)) 
test.default = test.default.map(dict(yes=1, no=0)) 
test.housing = test.housing.map(dict(yes=1, no=0)) 
co=test.select_dtypes(include='object').columns 
test_dum = pd.get_dummies(test,columns=co)
X_test = test_dum 
```
